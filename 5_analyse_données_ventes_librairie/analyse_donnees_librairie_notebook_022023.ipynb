```python
# On commence par enlever les milisecondes de date qui ne nous seront pas utiles : 
date = df_transactions["date"].str.split(".", expand=True)
date.rename(columns = {0:"time", 1:"other"}, inplace = True)
date = date.drop("other", axis=1)
df_transactions = pd.concat([df_transactions, date], axis=1)
df_transactions = df_transactions.drop("date", axis=1)
df_transactions.rename(columns={'time': 'date'}, inplace=True)
df_transactions.head()
```

```python
#df_transactions["date"] = pd.to_datetime(df_transactions["date"], format="%Y-%m-%d %H:%M:%S")
```

En essayant de convertir au format date à partir de là, **on a un message d'erreur sur le format**. Il y a donc surement des dates ne correspondant pas au format normal. Regardons ça d'un peu plus près. On va travailler sur une copie de notre **`df_transactions`** pour essayer de localiser les erreurs possibles de notre variable `date`. On va appeler notre nouveau dataframe **`pbdate`**.

```python
# On va scinder la date en plusieurs variables différentes :
pbdate = df_transactions["date"].str.split("-", expand=True)
pbdate.rename(columns = {0:"year", 1:"month", 2:'date'}, inplace = True)
pbdate2 = pbdate["date"].str.split(" ", expand=True)
pbdate2.rename(columns = {0:"day", 1:"date"}, inplace = True)
pbdate3 = pbdate2["date"].str.split(":", expand=True)
pbdate3.rename(columns = {0:"hour", 1:"minute", 2:"second"}, inplace = True)
```

```python
# On merge pour retrouver notre dataframe initial avec la date scindée :
df_pbdate = pd.concat([df_transactions, pbdate, pbdate2, pbdate3], axis=1)
df_pbdate = df_pbdate.drop(["date"], axis=1)
df_pbdate
```

```python
# Regardons ce que contiennent nos variables de dates :
print("Années :", np.sort(pd.unique(df_pbdate["year"])))
print(" Mois", np.sort(pd.unique(df_pbdate["month"])))
print("Jours", np.sort(pd.unique(df_pbdate["day"])))
print("Heures", np.sort(pd.unique(df_pbdate["hour"])))
print("Minutes", np.sort(pd.unique(df_pbdate["minute"])))
print("Secondes", np.sort(pd.unique(df_pbdate["second"])))
```

```python
# Isolons ces individus dans un dataframe à part :
test_2021 = df_pbdate.loc[df_pbdate["year"] == "test_2021"]
test_2021
```

```python
print("id_prod :", np.sort(pd.unique(test_2021["id_prod"])))
print("session_id :", np.sort(pd.unique(test_2021["session_id"])))
print("client_id :", np.sort(pd.unique(test_2021["client_id"])))
```

On se rend compte avec le code ci-dessus que l'on a une année **"test_2021"**. Il n'y a pas d'autres soucis. C'est donc ce **"test_2021"** qui nous pose problème pour le passage au format to_datetime. En l'isolant on peut voir qu'on a **200 lignes différentes avec cette année test**. Nous allons dans un premier temps nettoyer notre **`df_transactions`**.

- Il faudrait regarder par la suite si les produits **"T_0"** ont uniquement des transactions test ou non. Si c'est le cas, nous pouvons supprimer ce produit test de notre dataframe **`produits`**.
- Il faudrait ensuite vérifier les clients **"ct_0" et "ct_1"**. Si ces clients ont uniquement effectué des transactions test alors une fois le produit test supprimé les clients ne devraient plus apparaitre, on peut donc les supprimer de notre dataframe **`clients`**.

```python
# On va isoler les transactions "test" :
T_0 = df_transactions.loc[df_transactions["id_prod"] == "T_0"]
T_0
```

On peut voir ici qu'on a le même nombre de lignes que dans **`test_2021`**, on a donc localisé toutes les transactions sur le produit test T_0. On peut supprimer les transactions, et supprimer le produit du dataframe **`produits`** par la suite.

Regardons ensuite s'il reste des transactions pour les "ct_0" et "ct_1".

```python
# On supprime les 200 lignes avec la date test :
df_transactions.drop(df_transactions[df_transactions["date"] == "test_2021-03-01 02:30:02"].index , inplace=True)
```

```python
# Vérifions s'il nous reste des lignes pour "ct_0" et "ct_1" :
ct = df_transactions.loc[(df_transactions["client_id"] == "ct_0") & (df_transactions["client_id"] == "ct_1")]
ct.shape
```

## 4.1. Chiffre d'affaires et nombre de ventes <a class="anchor" id="section_4.1."></a>

```python
# Création du dataframe df_ventes :
df_ventes = df_final2
```

### 4.1.1. CA Total et nombre de ventes  <a class="anchor" id="section_4.1.1."></a>

```python
# On peut calculer le chiffre d'affaires total :
ca_tot = round(df_ventes["price"].sum(), 2)
ventes_tot = df_ventes["session_id"].count()
print("Le chiffre d'affaires total est de", ca_tot, "€. Pour un nombre total de", ventes_tot,"ventes.")
```

### 4.1.2. CA annuel et nombre de ventes  <a class="anchor" id="section_4.1.2."></a>

```python
# On calcule le chiffre d'affaires réalisé par année :
df_ventes["year"] = df_ventes["date"].dt.year
ca_annuel = df_ventes.groupby(["year"])["price"].sum()
print("Chiffre d'affaires par année", ca_annuel)

# On peut aussi calculer le nombre de ventes par année :
ventes_annee = df_ventes.groupby(["year"])["session_id"].count()
print('*' *40)
print("Nombre de ventes par année", ventes_annee)
```

Notre **meilleure année a donc été 2022**, que ce soit pour notre chiffre d'affaires et pour nos nombres de ventes. Cependant 2023 n'est pas une année encore complétée, on peut donc faire une analyse mensuelle du chiffre d'affaires maintenant, cela nous permettra de voir si les ventes en début d'année 2023 sont meilleures que les années précédentes et que l'augmentation de nos ventes se confirme.

### 4.1.3. CA mensuel et nombre de ventes  <a class="anchor" id="section_4.1.3."></a>

```python
# On commence par créer une variable année-mois :
df_ventes["mois_annee"] = df_ventes["date"].dt.strftime("%Y-%m")
df_ventes = df_ventes.drop(["year"], axis=1)
```

```python
# On créé ensuite un dataframe en goupant les mois :
ca_evolution = df_ventes.groupby(["mois_annee"], as_index=False)["price"].sum()
ca_evolution = ca_evolution.sort_values(by=["mois_annee"])
```

Afin de rendre les choses plus visuelle ici au vue du grand nombre de mois que nous avons à faire apparaitre, nous allons représenter tout çà sur un diagramme à batons.

```python
# Diagramme à bâtons montrant l'évolution du chiffre d'affaires dans le temps :
plt.figure(figsize=(24,10))
sns.barplot(y=ca_evolution['price'], x=ca_evolution["mois_annee"], color="gray")
plt.xlabel("")
plt.ylabel("Montants (€)")
plt.title("Evolution mensuelle du chiffre d'affaires", fontsize = 18, fontweight="bold")
plt.show()
```

Le diagramme ci-dessus fait clairement ressortir une **baisse du chiffre d'affaires en octobre 2021**.

## 4.2. Anomalie du mois d'octobre 2021 <a class="anchor" id="section_4.2."></a>

Dans cette sous-partie on va essayer de comprendre la baisse du chiffre d'affaires constatée pour le mois d'octobre 2021. On va donc analyser notre chiffre d'affaires et nos ventes par catégorie de produit pour voir si le problème est général ou uniquement lié à une seule catégorie de produit.

```python
# On va créer un dataframe spécial par catégorie :
df_analyse_oct = df_ventes[["categ", "price", "mois_annee"]]
df_analyse_oct.set_index('mois_annee', inplace=True)
df_analyse_oct = pd.pivot_table(df_analyse_oct, index=["mois_annee"], columns=["categ"], aggfunc=["sum", "count"]).reset_index()
df_analyse_oct.columns = ["Date", "categ_0", "categ_1", "categ_2", "vol_categ_0", "vol_categ_1", "vol_categ_2"]
df_analyse_oct.head(10)
```

```python
# Liste de couleurs pastels pour pouvoir mieux interpréter les graphiques :
couleurs = ["#a6cee3", "#fdbf6f", "#676283", "#b2df8a", "#cab2d6", "#ff7f00"]
```

```python
# Graphique montrant le chiffre d'affaires par catégorie par mois:
df_analyse_oct[["categ_0", "categ_1", "categ_2"]].plot.bar(figsize=(14,6), color =couleurs, edgecolor = "grey", linewidth = 1)
plt.legend(["catégorie 0", "catégorie 1", "catégorie 2"])
plt.xlabel("Mois")
plt.ylabel("Montants (en €)")
plt.title("Evolution mensuelle du chiffre d'affaires par catégorie", fontsize = 14, fontweight="bold")
```

```python
# Graphique montrant le nombre de ventes par catégorie par mois :
df_analyse_oct[["vol_categ_0", "vol_categ_1", "vol_categ_2"]].plot.bar(figsize=(14,6), color =couleurs, edgecolor = "grey", linewidth = 1)
plt.legend(["catégorie 0", "catégorie 1", "catégorie 2"])
plt.xlabel("Mois")
plt.ylabel("Nombre de ventes")
plt.title("Evolution mensuelle du nombre de ventes par catégorie", fontsize = 14, fontweight="bold")
```

Ces deux graphiques nous permettent de pouvoir analyser un peu plus en détail le mois d'octobre 2021. Tout d'abord **on repère assez rapidement que la baisse du chiffre d'affaires et des ventes est dû à la baisse du nombre de vente de la catégorie 1**. ### 7.1.1. Courbe de Lorenz  <a class="anchor" id="section_7.1.1."></a>

La courbe de Lorenz est un outil graphique utilisé en économie pour représenter la distribution de la richesse ou du revenu dans une population. Elle permet de visualiser l'écart entre la distribution théorique égale de la richesse ou du revenu et la distribution réelle.

```python
# Courbe de Lorentz : (repartition du CA par clients)
lorenz = np.cumsum(np.sort(meilleurs_clients)) / meilleurs_clients.sum()
lorenz = np.append([0], lorenz)
plt.figure(figsize=(14,8))
plt.plot(np.linspace(0,1,len(lorenz)), lorenz, drawstyle="steps-post",label = "Courbe de Lorenz")
plt.plot(np.linspace(0,1,len(lorenz)), np.linspace(0,1,len(lorenz)), color ="grey", dashes=[2,4], label = "Première bissectrice")
plt.ylabel("Part cumulée du chiffre d'affaires")
plt.xlabel("Part cumulée des clients")
plt.title("Répartition du chiffre d'affaires entre les clients", fontsize = 15, fontweight="bold")
plt.legend(loc='upper left', fontsize=10)

# Ajout d'une ligne de croisement de la courbe de Lorenz :
vline_x = 0.8 # la valeur x où vous souhaitez placer la ligne verticale
vline_y = np.interp(vline_x, np.linspace(0,1,len(lorenz)), lorenz)
plt.plot([0, vline_x], [vline_y, vline_y], color='red', linestyle='--', linewidth=1)
plt.plot([vline_x, vline_x], [0, vline_y], color='red', linestyle='--', linewidth=1)

# Calcul de l'indice de Gini :
aire_ss_courbe_CA_client = lorenz[:-1].sum() / len(lorenz)
S_ca_client = 0.5 - aire_ss_courbe_CA_client
gini_CA_client = 2*S_ca_client
print("indice de Gini:",round(gini_CA_client,3))
```

L'indice de Gini mesure le degré d'inégalité de la répartition de la variable étudiée entre les individus de la population. Sa valeur est comprise entre 0 et 1, où 0 correspond à une répartition parfaitement égale et 1 correspond à une répartition totalement inégale (un seul individu détient toute la variable étudiée).


On peut remarquer avec ce graphique que **le chiffres d'affaires est concentré sur quelques clients**. Le reste du chiffre d'affaires est plutôt bien réparti entre les clients. Essayons de refaire cette courbe de Lorenz en excluant les 4 plus gros clients, qui concentrent une grande part du chiffre d'affaires.

Pour interpréter ce graphique on peut dire que :

- **50 % du chiffre d'affaires est réalisé par 80 % des clients qui dépensent le moins.**
- **les 50 % de chiffre d'affaires restant sont réalisés par les 20 % des plus gros clients.**

## 7.2. Lien entre le genre d’un client et les catégories des livres achetés <a class="anchor" id="section_7.2."></a>

Nous avons supprimer les 4 plus gros clients de la suite de nos analyses car se sont des clients professionnels qui pourraient fausser nos résultats.

Le code ci-dessous est utilisé pour effectuer un **test du khi-deux** qui va évaluer si **deux variables qualitatives** sont liées. Dans le test qui nous intéresse l’hypothèse nulle est simplement que les deux variables testées (à savoir le sexe et la catégorie des livres achetés) sont indépendantes.

Y a-t-il une corrélation entre le sexe des clients et les catégories de produits achetés ?

Pour ce test on calcul un tableau appelé `indep` qui représente la distribution attendue en cas d'indépendance entre les deux variables. Le `tableau_contingence` est rempli avec les données et les valeurs manquantes sont remplacées par des zéros. Le tableau `measure` est alors calculé à partir de ces deux tableaux, et représente les écarts entre les observations et les valeurs attendues sous l'hypothèse d'indépendance.

La variable `xi_n` est la statistique du test khi-deux, qui est la somme de tous les écarts au carré normalisée par le nombre total d'observations.

### 7.2.1. Création d'un tableau de contingence  <a class="anchor" id="section_7.2.1."></a>

```python
# On commence par faire un tableau de contingence :
X = "sex"
Y = "categ"

tableau_contingence = df_stats[[X,Y]].pivot_table(index=X, columns=Y, aggfunc=len, margins=True, margins_name = "Total")
tableau_contingence
```

Une rapide analyse du tableau de contingence ci-dessus nous permet de constater que les ventes de chacune des catégories sont supérieures chez les femmes.

Il n'est pas nécessaire de faire un test de normalité avant de faire un test du khi-deux entre deux variables qualitatives, car le test du khi-deux est un test non-paramétrique qui ne repose pas sur une hypothèse de normalité.

### 7.2.2. Test du khi-deux  <a class="anchor" id="section_7.2.2."></a>

On va émettre les deux hypothèses suivantes :
- **H0 : Les deux variables sont indépendantes**
- **H1 : Les deux variables ne sont pas indépendantes**

$\alpha = 5\%$

```python
# On peut maintenant calculer le khi-deux :
chi2, pvalue, degrees, expected = st.chi2_contingency(tableau_contingence)
```

```python
print(chi2)
print(pvalue)
print(degrees)
```

### 7.3.3. Test de Spearman  <a class="anchor" id="section_7.3.3."></a>

```python
from scipy.stats import spearmanr

# Calcul du coefficient de corrélation de Spearman et du p-value
corr, p_value = spearmanr((age_ca[["age"]]), (age_ca["sum"]))

# Affichage des résultats
print("Coefficient de corrélation de Spearman :", corr)
print("P-value :", p_value)
```

Le test de corrélation de Spearman permet de tester l'existence d'une relation monotone entre deux variables. La corrélation de Spearman est comprise entre -1 et 1, et une valeur proche de 1 indique une forte corrélation positive, **une valeur proche de -1 indique une forte corrélation négative**, et une valeur proche de 0 indique une absence de corrélation.

Le test de Spearman est également associé à une `p-value`, qui indique la probabilité d'obtenir une corrélation aussi élevée ou plus élevée si les deux variables sont indépendantes (c'est-à-dire si la corrélation est nulle). **Ici la `p-value` < 0,05, on considère que la corrélation est significative.**

L'**ANOVA** (Analyse de Variance) est une méthode statistique qui permet de comparer les moyennes de plusieurs groupes. C'est un test statistique paramétrique qui peut être utilisé pour tester l'hypothèse nulle selon laquelle les moyennes de tous les groupes sont égales.

On va donc **vérifier si nos échantillons suivent une loi normale, puis si nos variances sont égales**. Si les variances pour chaque groupe ne sont pas égales, on utilisera le **test de Kruskal – Wallis**.

### Test de Kolmogorov-Smirnov - vérification de la normalité de chaque groupe de la variable 'categ'

Le test de Kolmogorov-Smirnov est un test statistique non paramétrique utilisé pour déterminer si une distribution de données suit une distribution théorique connue.

Avec un risque premier $\alpha = 5\%$, on émets les deux hypothèses suivantes :

- **H0 : L'échantillon suit une distribution normale.**
- **H1 : L'échantillon ne suit pas une distribution normale.**

```python
from scipy.stats import kstest, norm

# Les 3 échantillons à tester :
ca_0 = age_categ[age_categ['categ'] == 0]['age']
ca_1= age_categ[age_categ['categ'] == 1]['age']
ca_2 = age_categ[age_categ['categ'] == 2]['age']

alpha = 0.05 #niveau de significativité

# Test de normalité pour ca_0 :
kstat, pvalue = kstest(ca_0, norm.fit(ca_0))

if pvalue < alpha:
    print("La catégorie 0 ne suit pas une distribution normale")
else:
    print("La catégorie 0 suit une distribution normale")
print("p-value =", pvalue)

# Test de normalité pour ca_1 :
kstat, pvalue = kstest(ca_1, norm.fit(ca_1))

if pvalue < alpha:
    print("La catégorie 1 ne suit pas une distribution normale")
else:
    print("La catégorie 1 suit une distribution normale")
print("p-value =", pvalue)

# Test de normalité pour ca_2 :
kstat, pvalue = kstest(ca_2, norm.fit(ca_2))

if pvalue < alpha:
    print("La catégorie 2 ne suit pas une distribution normale")
else:
    print("La catégorie 2 suit une distribution normale")
print("p-value =", pvalue)
```

Nos échantillons suivent donc une loi normale, testons maintenant nos variances pour voir si elles sont égales ou non.

### Test de Levene - vérification d'égalité des variances entre les échantillons

Le test de Levene est un test statistique paramétrique utilisé pour évaluer l'homogénéité de la variance entre les groupes d'une variable continue. **Le test de Levene suppose que les données suivent une distribution normale** ou approximativement normale, ce qui en fait un **test paramétrique**.

```python
# Test d'égalité des variances entre les échantillons :

from scipy.stats import levene

stat, pvalue = levene(ca_0, ca_1, ca_2)

if pvalue < alpha:
    print("Les variances ne sont pas égales.")
else:
    print("Les variances sont égales.")
```

**Nos variances ne sont pas égales**, nous allons donc utiliser **le test de Kruskal-Wallis** à la place d'ANOVA.

Nous aurions pu faire un test de Student pour comparer les moyennes des échantillons. Cependant ce test est généralement utilisé dans les cas où les données suivent une distribution normale et que les variances sont approximativement égales, ce qui n'est pas notre cas.


### Test de Kruskal-Wallis

Le **test de Kruskal-Wallis** est une alternative à l'ANOVA qui peut être utilisée lorsque les conditions de l'ANOVA ne sont pas remplies, notamment lorsque les variances ne sont pas égales. Le test de Kruskal-Wallis est un test non paramétrique qui permet de comparer les médianes de plusieurs groupes. Il permet de faire un test entre **une variable qualitative et une variable quantitative.**

```python
# Test de Kruskal-Wallis :

from scipy.stats import kruskal

k_stat, p_value = kruskal(ca_0, ca_1, ca_2)

alpha = 0.05 #niveau de significativité

# Interprétation du test :
if p_value < alpha:
    print("Il y a une différence statistiquement significative entre les groupes.")
else:
    print("Il n'y a pas de différence statistiquement significative entre les groupes.")
```

Ce projet a permis de faire un état des lieux de la librairie en ligne après deux ans d'existence. Les données des différents dataframes étaient de bonne qualité dans l'ensemble, mais le format de la date devait être corrigé.

L'analyse du chiffre d'affaires a permis de repérer une anomalie pour le mois d'octobre 2021. En général, le chiffre d'affaires global est resté plutôt stable dans le temps. Les produits des catégories 0 et 1 sont les plus vendus et également les plus représentés dans notre catalogue. Nous avons constaté que les produits les moins chers sont les plus représentés dans chacune de nos catégories.

En ce qui concerne les clients, l'âge moyen est de 45 ans et la tranche d'âge la plus représentée est celle des 35-45 ans. Nous avons légèrement plus de femmes que d'hommes. Nous avons également remarqué quatre gros clients qui pourraient être des grossistes, et ils représentent une part importante de notre chiffre d'affaires. Nous avons constaté une corrélation entre l'âge et le genre des clients et les catégories de livres achetés, entre l'âge des clients et le montant total des achats, ainsi qu'entre l'âge des clients et la fréquence d'achat, mais pas nécessairement entre l'âge des clients et la taille du panier moyen (plusieurs groupes ressortent).
